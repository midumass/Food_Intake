```python
import pandas as pd
import numpy as np
```

```python
food = pd.read_csv('./food_intake.csv')
```

```python
food.head(3)
```

```python
# We only need the percent change from 1975 to present
obs = food[food.Obesity.notnull()]
```

```python
obs.head()
```

```python
# For entries not recorded in 1975, values are backfilled from the most recent data
obs = obs.fillna(method = 'bfill')
```

```python
# Rates of change for categories to identify which are worth including
obs_chg = obs.query('Year == 2017 or Year == 1975').pct_change(fill_method = 'ffill')
```

```python
obs_chg.head()
```

```python
# Change rates to percentages
obs_pct = obs_chg.iloc[[1]] * 100

obs_pct
```

```python
# Which items have decreased in consumption by more than 25% or increased in consumption by more than double?
obs_pct.loc[:, ((obs_pct > 100) | (obs_pct < -25)).any(axis = 0)].T
```

```python
obs_pct = obs_pct.T

obs_pct
```

```python
# Lets rename that '47' to values
obs_pct = obs_pct.rename({47:"values"}, axis = 'columns')
```

```python
obs_pct.sort_values(by = 'values', ).head(10)
```

```python
# We see the items that stopped being recorded here in the top 10, so lets drop them and recheck.
obs_pct.drop(['Italian Cheese Parmesan',\
              'Italian Cheese Ricotta',\
              'Italian Cheese Provolone',\
              'Durum flour',\
              'Italian Cheese Romano']).sort_values(by = 'values').head(50)
```

```python
# Lets see the other side.
obs_pct.sort_values(by = 'values', ascending = False).head(10)
```

```python
# Again some bad values. Other italian cheese has a normalized column, making this one irrelevant. Obesity also can go.
obs_pct.drop(['Other Italian cheese', 'Obesity']).sort_values(by = 'values', ascending = False).head(50)
```
